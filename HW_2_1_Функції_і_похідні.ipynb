1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sp.symbols('x')
f = 10*x - 5
f_derivative = sp.diff(f, x)
print("Похідна лінійної функції 10x-5:", f_derivative)
```

```python
xs = np.linspace(0,10,40)
plt.plot(xs, f(xs), 'o');
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_1 = sp.sin(x)
f_2 = sp.ln(x)
f_1_derivative = sp.diff(f_1, x)
f_2_derivative = sp.diff(f_2, x)
print("Похідна функції sin(x):", f_1_derivative)
print("Похідна функції ln(x)", f_2_derivative)
```

```python
f1_func = sp.lambdify(x, f_1, 'numpy')
f1_prime_func = sp.lambdify(x, f_1_derivative, 'numpy')
f2_func = sp.lambdify(x, f_2, 'numpy')
f2_prime_func = sp.lambdify(x, f_2_derivative, 'numpy')

x_vals = np.linspace(0.1, 10, 500)
y1_vals = f1_func(x_vals)
y_1_derivative_vals = f1_prime_func(x_vals)
y2_vals = f2_func(x_vals)
y2_derivative_vals = f2_prime_func(x_vals)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x_vals, y1_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, y_1_derivative_vals, label=f"f'(x) = cos(x)", color='red', linestyle='--')
plt.title('Графіки функції f(x) = sin(x) та її похідної')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x_vals, y2_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, y2_derivative_vals, label=f"f'(x) = 1/x", color='red', linestyle='--')
plt.title('Графіки функції f(x) = ln(x) та її похідної')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f_3 = x**2 - 5*x + 1
f_3_derivative = sp.diff(f_3, x)
print("Похідна квадратної функції x**2 - 5*x + 1:", f_3_derivative)
```

```python
f3_func = sp.lambdify(x, f_3, 'numpy')
f3_prime_func = sp.lambdify(x, f_3_derivative, 'numpy')

x_vals = np.linspace(0, 100, 500)
y3_vals = f3_func(x_vals)
y_3_derivative_vals = f3_prime_func(x_vals)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x_vals, y3_vals, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_vals, y_3_derivative_vals, label=f"f'(x) = 2x-5", color='red', linestyle='--')
plt.title('Графіки функції f(x) =  x^2 - 5x + 1 та її похідної')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
